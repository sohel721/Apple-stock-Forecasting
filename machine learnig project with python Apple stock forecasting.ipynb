```python
import pandas as  pd
```

```python
df = pd.read_csv("F:\Data analytics\\AAPL.csv")
```

```python
df.head()
```

```python
import matplotlib.pyplot as plt
```

```python
df = df[['Date' , 'Close']]
df['Date'] = pd.to_datetime(df.Date)
```

```python
#df['Date']
```

```python
df['Close'].plot(figsize = (12,8) , title = "Apple stock prizes" , fontsize = 15 , label = "close prize")
plt.legend()
plt.grid()
plt.show()
```

```python
from autots import AutoTS
```

```python
model = AutoTS(forecast_length = 10 , frequency = 'infer' ,
              ensemble = 'simpe' , drop_data_older_than_periods = 200)
model = model.fit(df , date_col = 'Date' , value_col = 'Close' , id_col = None)                
```

```python
prediction = model.predict()
forecast = prediction.forecast
print("Stock  prize prediction of Apple")
print(forecast)
```
